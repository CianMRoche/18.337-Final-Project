```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of datapoints list
N = [1000,10000,100000]


# it/s lists
serial = [33000,3125,330]
threads_4 = [130000,14600,1590]
threads_12 = [333000,40000,4160]    # maybe do relative to serial
```

```python
plt.plot(N,np.divide(threads_12,serial),'s',color='red',label='Julia 12 threads')
plt.plot(N,np.divide(threads_4,serial),'^',color='C1',label='Julia 4 threads')

plt.title('Performance scaling with number of threads')
plt.xlabel(r'N')
plt.ylabel('iterations per second relative to serial')
#plt.yscale('log')
plt.xscale('log')
plt.grid()
plt.legend()
```
