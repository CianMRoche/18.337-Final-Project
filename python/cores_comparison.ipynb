```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Number of datapoints list
num_cores = np.array([1,2,3,4,6,8,16,32,64,128])


# it/s lists
walkers_5 = np.array([25,40,60,82,90,90,90,90,90,90])*5
walkers_50 = np.array([2,4,9,13,19,30,44,80,80,80])*50    # maybe do relative to serial
```

```python
plt.plot(num_cores,walkers_50,'s',color='cyan',label='Python 50 walkers')
plt.plot(num_cores,walkers_5,'^',color='C0',label='Python 5 walkers')

plt.title('Affine-Invariant parallel scaling')
plt.xlabel(r'CPU cores')
plt.ylabel('Total iterations per second')
plt.yscale('log')
plt.xscale('log')
plt.grid()
plt.legend()
```
